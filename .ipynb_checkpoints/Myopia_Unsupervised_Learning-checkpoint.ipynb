# Unsupervised Machine Learning
In this notebook, I will be running unsupervised machine learning on a data set of patient data with regards to myopia. First I will prepare the data by removing a target column, in order to make sure the machine learning is correct and effective. Then I will apply dimensionality reduction to standardize how the data is perceived. Next, I will perform cluster analysis, and finally make a recommendation on the model.

## Prepare Data

```python
# Initial imports
import pandas as pd
from pathlib import Path
```

```python
# Data loading
file_path = Path("myopia.csv")
df = pd.read_csv(file_path)
df.head(10)
```

```python
# Find null values
for column in df.columns:
    print(f"Column {column} has {df[column].isnull().sum()} null values")
```

```python
# Drop MYOPIC column and save
learn_df = df.drop(['MYOPIC'], axis = 'columns')
labels = df['MYOPIC']
learn_df.head(10)
```

```python
# Standardize data
from sklearn.preprocessing import StandardScaler
df_scaled = StandardScaler().fit_transform(learn_df)
print(df_scaled[0:5])
```

## Apply Dimensionality with PCA

```python
# Perform PCA
from sklearn.decomposition import PCA
pca = PCA(n_components = .9)
```

```python
# Fit and transform scaled with PCA
myopia_pca = pca.fit_transform(df_scaled)
myopia_pca
```

```python
df_myopia_pca = pd.DataFrame(data = myopia_pca)
df_myopia_pca.head(10)
```

```python
# Reduce dataset with t-SNE
from sklearn.manifold import TSNE

# Initialize t-SNE
tsne = TSNE(learning_rate=7)

# Reduce dimensions
tsne_features = tsne.fit_transform(myopia_pca)

tsne_features
```

```python
# View tsne_features for the shape
tsne_features.shape
```

```python
# Prepare to plot the dataset
x_values = tsne_features[:,0]

y_values = tsne_features[:,1]
```

```python
# Visualize the clusters with color
import matplotlib.pyplot as plt

plt.scatter(x_values, y_values, c=labels)
plt.show()
```

Clusters are forming, but nothing necessarily indicative of groupings

## Perform Cluster Analysis

```python
from sklearn.cluster import KMeans
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=7)
    km.fit(myopia_pca)
    inertia.append(km.inertia_)

# Create the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.head()
```

```python
# Plot elbow curve to find the best number for K
plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

## Recommendation

Based on the model as presented here, there are three clusters forming in the data after t-SNE reduction. This is also borne out in the the K-mean model as the sharpest angle exists at 3.
